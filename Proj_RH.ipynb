```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Base_Salario = pd.read_csv('Salary_Data.csv')

Base_Salario.head()
```

```python
Base_Salario.info()
```

```python
Base_Salario.shape
```

```python
Base_Salario.columns
```

**'inplace = True'** - A modificação será aplicada no dtframe original, se fosse 'false' seria criado outro dtframe

```python
Base_Salario.rename(columns={
    'YearsExperience' : 'Anos de Experiencia',
    'Salary' : 'Salario'
    }, inplace= True)
```

```python
Base_Salario.head()
```

```python
# Verificando e somando os campos nulos, se existirem
Base_Salario.isnull().sum()
```

Gráfico dos valores nulos

```python
sns.heatmap(Base_Salario.isnull(), cbar = False)
# OU SÓ 'sns.heatmap(Base_Salario.isnull())'
```

```python
Base_Salario.describe()
```

```python
plt.figure(figsize= (8,4))
plt.title('Análise de Renda')
sns.kdeplot(Base_Salario['Salario'], color= 'green');
```

```python
plt.figure(figsize= (8,4))
plt.title('Análise de Experiencia')
sns.kdeplot(Base_Salario['Anos de Experiencia'], color= 'blue');
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(Base_Salario['Salario'])
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(Base_Salario['Anos de Experiencia'])
```

```python
plt.figure(figsize=(8,4))
plt.title('Anos de Experiencia x Sálario')
sns.scatterplot(data = Base_Salario, x= 'Anos de Experiencia', y= 'Salario')
```

```python
plt.figure(figsize=(8, 4))
plt.title('Análise de Teste')
sns.regplot(data=Base_Salario, x='Anos de Experiencia', y= 'Salario')
```

```python
plt.figure(figsize= (8,4))
plt.title('Analise de Teste')
sns.regplot(data= Base_Salario, x= 'Salario', y ='Anos de Experiencia')
```
