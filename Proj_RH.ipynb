### **Gráfico Kdeplot**
Mostra a distribuição de uma unica variável, de acordo com seus valores. Sendo assim, mostrará:

- Quais valores da variável aparecem mais frequentemente (onde a linha está mais alta).
- Quais valores aparecem com menos frequência (onde a linha está mais baixa).
- Como esses valores se espalham (se a linha é larga ou estreita)

```python
plt.figure(figsize= (8,4))
plt.title('Análise Referente aos Salários', loc= 'left', fontweight= 'bold', fontsize= 12)
sns.kdeplot(Base_Salario['Salario'], color= '#CD5C5C')
plt.xlabel('Salário')
plt.ylabel('Distribuição')
```

```python
plt.figure(figsize= (8,4))
plt.title('Análise Referente aos Anos de Experiência', loc= 'left', fontweight= 'bold', fontsize= 12)
sns.kdeplot(Base_Salario['Anos de Experiencia'], color= '#2E8B57')
plt.xlabel('Experiência')
plt.ylabel('Distribuição')

```

```python
plt.figure(figsize=(8,6))
plt.title('Análise da Distribuição dos Salários', loc= 'left', fontweight= 'bold', fontsize= 12, color= '#4682B4' )
sns.boxplot(Base_Salario['Salario'])
plt.ylabel('Salário', fontweight = 600)
```

```python
plt.figure(figsize=(8,6))
plt.title('Análise da Distribuição dos Anos de Experiência', loc='left', fontweight= 'bold', color= '#2E8B57')
sns.boxplot(Base_Salario['Anos de Experiencia'], color= '#2E8B57')
plt.ylabel('Anos de Experiência', fontweight= 600)
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data = Base_Salario, x= 'Anos de Experiencia', y= 'Salario')
plt.title('Análise do Comparativo de Anos de Experiência e Salário', loc= 'left', fontweight = 'bold', color = '#4682B4')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
```

```python
plt.figure(figsize=(8, 4))
sns.regplot(data=Base_Salario, x='Anos de Experiencia', y= 'Salario')
plt.title('Análise do Comparativo de Anos de Experiência e Salário', loc= 'left', fontweight = 'bold', color = '#4682B4')
```
