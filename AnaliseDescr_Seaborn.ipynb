```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
Base_Dados = sns.load_dataset('iris')

Base_Dados.head()
```

```python
Base_Dados.shape
```

Medidas de Tendencia Central - indicam um ponto em torno do qual se concentram os dados

```python
#Média (+/)
Base_Dados["petal_length"].mean()
```

```python
#Moda
Base_Dados["petal_length"].mode()
```

```python
#Mediana (meio)
Base_Dados["petal_length"].median()
```

Medidas Separatrizes -  ocupam posições no conjunto de dados, dividindo em partes iguais.

1 quartil - Q1 25%
2 quartil - Q2 = M4 50%
3 quartil = Q3 75%

```python
Base_Dados["petal_length"].describe()
```

```python
'''
25% - 1.0 --> 1.6
50% - 1.6 --> 4.3
70% - 4.3 --> 5.1
100 - 5.1 --> 6.9
'''
```

```python
sns.boxplot(Base_Dados['petal_length'])
```

Medidas de Dispersão - indicam se os dados estão ou não proximos um do outro. Auxiliam as medidas de tendencia.

```python
Base_Dados['sepal_length'].unique()
```

```python
Base_Dados['sepal_length'].describe()
```

```python
# Amplitude Total (diferença entre o maior e menor valor)
Base_Dados['sepal_length'].max() - Base_Dados['sepal_length'].min()
```

```python
# Amplitude Interquartilica (diferença entre o terceiro quartil (75%) e o primeiro quartil (25%), medida mais estavel que a amplitude total, por não considerar valores extremos)
Base_Dados['sepal_length'].describe()[6:7].values - Base_Dados['sepal_length'].describe()[4:5].values
```

```python
# Amplitude Semi-Interquartilica (diferença entre o terceiro quartil (75%) e o primeiro quartil (25%), dividido por 2)
Base_Dados['sepal_length'].describe()[6:7].values - Base_Dados['sepal_length'].describe()[4:5].values/ 2
```

```python
# Variancia (mostra o quão distante os valores estão da média)
Base_Dados['sepal_length'].var()
```

```python
# Desvio Padrão (o resultado positivo da raiz quadrada da variancia, quanto mais próximo de 0, mais os dados estão proximos da média.)
Base_Dados['sepal_length'].std()
```

Medidas de Assimetria - indicador da forma de distribuição dos dados. Ao construir uma distruição de tendencia e/ou histograma, está buscando identificar a distruição dos dados

```python
'''
Simetria: média = mediana = moda ( = 0)
Assimétrica negativa: média < mediana < moda ( < 0)
Assimetria positiva: moda < mediana < média ( > 0)
'''
```

```python
# Calcular a simetria
Base_Dados['sepal_width'].skew()
```

```python
sns.kdeplot(Base_Dados['sepal_length']);
```

Medida de Curtose - é o grau de achatamento da distruição:

Leptocúrtica - curva de frequencia mais fechada, k < 0,263
Mesocúrtica - dados razoalmente concentrados em torno do centro, k = 0,263
Platicúrtica - curva de frequencia mais aberta, k > 0,263

```python
# Medir a curtose
Base_Dados['sepal_length'].kurtosis()
```

```python
Base_Dados['sepal_length'].kurtosis()
```

Correlação - quando ocorre a hipótese de que o aumento de uma variavel ou queda está associado à evolução de outra variável

Correlação de Pearson - tem um intervalo de 1 a -1.

Se der 0  - não há correlação

Associação positiva - valor > 0 (as duas variaveis aumentam)

Associação negativa - valor < 0 (a medida que uma aumenta, a outra diminui):

```python
Base_Dados.select_dtypes(include=['number']).corr()
```

```python
sns.heatmap(Base_Dados.select_dtypes(include=['number']).corr(), annot=True)
```

```python
sns.scatterplot(data=Base_Dados, x= 'petal_length', y= 'petal_width') #comprimento da pétala, largura da pétala
```

Correlação de Sperman - Uma vez que a correlação de Spearman segue uma lógica monotética, e não tem pressupostos lineares como na correlação de Pearson, é possível utilizar o rs para relações não lineares

```python
Base_Dados.select_dtypes(include=['number']).corr('spearman')
```

```python
sns.heatmap( Base_Dados.select_dtypes(include=['number']).corr( 'spearman' ), annot=True )
```

```python
sns.scatterplot( data=Base_Dados, x='sepal_length', y='petal_width')
```
